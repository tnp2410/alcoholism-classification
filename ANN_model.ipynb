```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

```python
train = pd.read_csv('data/train_df.csv')
test = pd.read_csv('data/test_df.csv')
validation_subjects = ['co2a0000377', 'co2a0000364', 'co2c0000342','co2c0000345']
validation = test[test['subject'].isin(validation_subjects)].drop(columns=['sample', 'match', 'err', 'object'])
test = test[~test['subject'].isin(validation_subjects)]
train = train.drop(columns=['sample', 'match', 'err', 'object'])
test = test.drop(columns=['sample', 'match', 'err', 'object'])
#data = pd.concat([train.drop(columns=['sample', 'match', 'err', 'object', 'subject']), test.drop(columns=['sample', 'match', 'err', 'object', 'subject'])])


x_train = train.drop(columns=['alcoholic', 'subject'])
y_train = train[['alcoholic']]

x_test = test.drop(columns=['alcoholic', 'subject'])
y_test = test[['alcoholic']]

x_validation = validation.drop(columns=['alcoholic', 'subject'])
y_validation = validation[['alcoholic']]

#x_train, x_test, y_train, y_test = train_test_split(data.drop(columns=['alcoholic']), data[['alcoholic']], train_size=.8, test_size=.2)
pca = PCA(n_components=30)

x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
x_validation = pca.transform(x_validation)

model = Sequential()
model.add(Dense(50, input_dim=30, activation='tanh'))
model.add(Dense(1, activation='sigmoid')) 
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=100, batch_size=16, validation_data=(x_test, y_test))
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
